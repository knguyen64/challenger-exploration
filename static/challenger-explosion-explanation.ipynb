```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

## Data

```python
# Read Data
file = 'o-ring-erosion-or-blowby.data'
data = pd.read_csv(file, header=None).apply(lambda x: x.str.split())

# Clean Data
data = pd.DataFrame(data[0].to_list()).rename(columns={
    0:'o_rings', 
    1:'o_rings_thermal_stress', 
    2:'launch_temp', 
    3:'leak_check_pressure', 
    4:'temporal_order'})
data['outcome'] = data['o_rings_thermal_stress'].apply(lambda x: 0 if int(x)==0 else 1)
data
```

## Logistic Regression Model

```python
# Fitting Logistic Regression Model

clf = LogisticRegression(max_iter=10000)
clf.fit(data[['launch_temp']], data['outcome'])
coef = clf.coef_[0][0]
intercept = clf.intercept_[0]
coef, intercept
```

```python
def prob_explosion(temp):
    return np.exp(intercept + coef*temp)/(1+np.exp(intercept + coef*temp))

prob = prob_explosion(36)
prob
```
